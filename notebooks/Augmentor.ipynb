```python
import cv2 as cv 
import os 
import Augmentor
```

```python
# PATH for the citizenship , passport and license 
input_dir = "data/processed/citizenship"
input_dir1 = "data/processed/passport"
input_dir2 = "data/processed/license"
```

```python
# Path for the Citizenship , passport and license 
output_dir = "../../augmented/citizenship"
output_dir1 = "../../augmented/passport"
output_dir2= "../../augmented/license"
```

Note : If there are Directories inside the path we provide then Augmentor will read 0 images , make sure that there are no subdirectories in the path where we are providing for the input 

```python
# Create separate output directories for each transformation
rotation_dir = os.path.join(output_dir, "rotation")
flip_dir = os.path.join(output_dir, "flip")
scaling_dir = os.path.join(output_dir, "scaling")
shear_dir = os.path.join(output_dir, "shear")
zoom_dir = os.path.join(output_dir, "zoom")
brightness_dir = os.path.join(output_dir, "brightness")
# noise_dir = os.path.join(output_dir, "noise")
# blur_dir = os.path.join(output_dir, "blur")
elastic_dir = os.path.join(output_dir, "elastic")
crop_dir = os.path.join(output_dir, "crop")
erase_dir = os.path.join(output_dir, "erase")

# Create Augmentor pipelines for each transformation type
rotation_pipeline = Augmentor.Pipeline(input_dir, rotation_dir)
flip_pipeline = Augmentor.Pipeline(input_dir, flip_dir)
scaling_pipeline = Augmentor.Pipeline(input_dir, scaling_dir)
shear_pipeline = Augmentor.Pipeline(input_dir, shear_dir)
zoom_pipeline = Augmentor.Pipeline(input_dir,zoom_dir)
brightness_pipeline = Augmentor.Pipeline(input_dir,brightness_dir)
# blur_pipeline = Augmentor.Pipeline(input_dir, blur_dir)
elastic_pipeline = Augmentor.Pipeline(input_dir, elastic_dir)
crop_pipeline = Augmentor.Pipeline(input_dir, crop_dir)
erase_pipeline = Augmentor.Pipeline(input_dir, erase_dir)

# Rotation Pipeline
rotation_pipeline.rotate(probability=1.0, max_left_rotation=25, max_right_rotation=0)

# Flip Pipeline
flip_pipeline.flip_left_right(probability=1.0)
flip_pipeline.flip_top_bottom(probability=1.0)

# Scaling Pipeline
scaling_pipeline.zoom_random(probability=1.0, percentage_area=0.8)
scaling_pipeline.random_contrast(probability=1.0, min_factor=0.5, max_factor=2.0)

# Shear Pipeline
shear_pipeline.shear(probability=1.0, max_shear_left=10, max_shear_right=10)

# Zoom Pipeline
zoom_pipeline.zoom_random(probability=1.0, percentage_area=0.8)

# Brightness Pipeline
brightness_pipeline.random_contrast(probability=1.0, min_factor=0.5, max_factor=2.0)
brightness_pipeline.random_brightness(probability=1.0, min_factor=0.7, max_factor=1.3)


# # Blur Pipeline
# blur_pipeline.blur(probability=0.5, min_sigma=0.5, max_sigma=1.5)

# Elastic Deformation Pipeline
elastic_pipeline.random_distortion(probability=1.0, grid_width=4, grid_height=4, magnitude=8)

# Random Cropping Pipeline
crop_pipeline.crop_random(probability=1.0, percentage_area=0.8)

# Random Erasing Pipeline
erase_pipeline.random_erasing(probability=1.0, rectangle_area=0.2)


# Process the images for each transformation type
rotation_pipeline.process()
flip_pipeline.process() 
scaling_pipeline.process() 
shear_pipeline.process()  
zoom_pipeline.process()  
brightness_pipeline.process() 
# noise_pipeline.process() 
# blur_pipeline.process()  
elastic_pipeline.process()  
crop_pipeline.process()  
erase_pipeline.process()  
```
